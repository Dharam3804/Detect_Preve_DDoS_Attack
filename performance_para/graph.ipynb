```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('data/TECH_perfor_para.csv')
data.head(20)
#print(data['Classifier'])
```

```python
x=data['Accuracy (%)']
#print(x)
x=x.ravel()
print(x)
```

```python
rf=[]
gbm=[]
svm=[]
knn=[]
st_EN=[]

for i in range(0,4):
    rf.append(x[i*5])
    gbm.append(x[i*5+1])
    svm.append(x[i*5+2])
    knn.append(x[i*5+3])
    st_EN.append(x[i*5+4])

print(rf)  
print(gbm) 
print(svm) 
print(knn) 
print(st_EN)   
```

```python
Groups = [rf,gbm,svm,knn,st_EN]

width = 0.8 / len(Groups)
lbl = ['RF','GBM','SVM','KNN','Stack Ensemble']
grp=['Information Gain','Chi-Squared','ReliefF','MAD']
Pos = np.array(range(len(Groups[0]))) 
A = ["A","B","C","D"]
plt.figure(figsize=(12,4))
for i in range(len(Groups)):
    plt.bar(Pos + i * width, Groups[i], width = width, edgecolor = 'white',label=lbl[i],align='center')
plt.xticks(Pos+0.3, grp)
plt.xlabel("Techniques",labelpad=20, fontweight ='bold', fontsize = 15)
plt.ylabel("Accuracy (%)",labelpad=20, fontweight ='bold', fontsize = 15)
plt.title("Accuracy comparision of various techniques alond with different classifier methods",pad=20)
plt.legend() 
plt.savefig('Accuracy.jpg',bbox_inches='tight')
plt.show()
```
